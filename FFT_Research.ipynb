# 正常週期性波型 FFT採樣

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# 設定參數
T = 1.0 / 800.0  # 取樣間隔
N = 800  # 取樣點數
x = np.linspace(0.0, N*T, N, endpoint=False)

# 不同類型的週期波
y_sine = np.sin(50.0 * 2.0*np.pi*x)  # 正弦波
y_square = np.sign(np.sin(50.0 * 2.0*np.pi*x))  # 方波
y_sawtooth = 2*(x*50.0 - np.floor(x*50.0 + 0.5))  # 鋸齒波

# 將不同波形組合在一起
y = y_sine + y_square + y_sawtooth

# 執行FFT轉換
yf = fft(y)
xf = fftfreq(N, T)[:N//2]

# 繪圖
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# 時域信號
axs[0].plot(x, y)
axs[0].set_title('Time Domain Signal')
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Amplitude')
axs[0].grid(linestyle = "--", color = "gray")

# 頻域信號
axs[1].plot(xf, 2.0/N * np.abs(yf[0:N//2]))
axs[1].set_title('Frequency Domain Signal')
axs[1].set_xlabel('Frequency [Hz]')
axs[1].set_ylabel('Amplitude')
axs[1].grid(linestyle = "--", color = "gray")

plt.tight_layout()
plt.show()
```

```python
# 增加取樣點數
N = 1600  # 取樣點數增加一倍
x = np.linspace(0.0, N*T, N, endpoint=False)

# 重新生成波形
y_sine = np.sin(50.0 * 2.0*np.pi*x)
y_square = np.sign(np.sin(50.0 * 2.0*np.pi*x))
y_sawtooth = 2*(x*50.0 - np.floor(x*50.0 + 0.5))
y = y_sine + y_square + y_sawtooth

# 執行FFT轉換
yf = fft(y)
xf = fftfreq(N, T)[:N//2]

# 繪圖
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# 時域信號
axs[0].plot(x, y)
axs[0].set_title('Time Domain Signal with Hamming Window')
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Amplitude')
axs[0].grid(linestyle = "--", color = "gray")

# 頻域信號
axs[1].plot(xf, 2.0/N * np.abs(yf[0:N//2]))
axs[1].set_title('Frequency Domain Signal with Hamming Window')
axs[1].set_xlabel('Frequency [Hz]')
axs[1].set_ylabel('Amplitude')
axs[1].grid(linestyle = "--", color = "gray")

plt.tight_layout()
plt.show()
```

# Stepwise Function FFT採樣

```python
# 增加取樣點數
N = 1600  # 取樣點數增加一倍
x = np.linspace(0.0, N*T, N, endpoint=False)

# 重新生成波形
y = np.random.normal(0, 0.01, size = (N, ))
I = np.arange(0, N, 100)
y[I] += 2
y[I] += np.random.normal(0, 0.01, size = (I.shape[0], ))
I = np.arange(0, N, 400)
y[I] -= 5
y[I] += np.random.normal(0, 0.01, size = (I.shape[0], ))
I = np.arange(0, N, 600)
y[I] += 10
y[I] += np.random.normal(0, 0.01, size = (I.shape[0], ))

# 應用Hamming窗口
window = np.hamming(N)
y = y * window

# 執行FFT轉換
yf = fft(y)
xf = fftfreq(N, T)[:N//2]

# 繪圖
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# 時域信號
axs[0].plot(x, y)
axs[0].set_title('Time Domain Signal with Hamming Window')
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Amplitude')
axs[0].grid(linestyle = "--", color = "gray")

# 頻域信號
axs[1].plot(xf, 2.0/N * np.abs(yf[0:N//2]))
axs[1].set_title('Frequency Domain Signal with Hamming Window')
axs[1].set_xlabel('Frequency [Hz]')
axs[1].set_ylabel('Amplitude')
axs[1].grid(linestyle = "--", color = "gray")

plt.tight_layout()
plt.show()
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

# 重新生成波形
y = np.random.normal(0, 0.01, size = (N, ))
I = np.arange(0, N, 100)
y[I] += 2
y[I] += np.random.normal(0, 0.01, size = (I.shape[0], ))
I = np.arange(0, N, 400)
y[I] -= 5
y[I] += np.random.normal(0, 0.01, size = (I.shape[0], ))
I = np.arange(0, N, 600)
y[I] += 10
y[I] += np.random.normal(0, 0.01, size = (I.shape[0], ))

# 應用移動平均濾波器，窗口大小為10
y_smooth = moving_average(y, 10)

# 執行FFT轉換
yf = fft(y_smooth)
xf = fftfreq(len(y_smooth), T)[:len(y_smooth)//2]

# 繪圖
fig, axs = plt.subplots(3, 1, figsize=(10, 9))

# (舊)時域信號
axs[0].plot(np.arange(len(y))*T, y)
axs[0].set_title('Original Time Domain Signal')
axs[0].set_xlabel('Time [s]')
axs[0].set_ylabel('Amplitude')
axs[0].grid(linestyle = "--", color = "gray")

# (新)時域信號
axs[1].plot(np.arange(len(y_smooth))*T, y_smooth)
axs[1].set_title('Smoothed Time Domain Signal')
axs[1].set_xlabel('Time [s]')
axs[1].set_ylabel('Amplitude')
axs[1].grid(linestyle = "--", color = "gray")

# 頻域信號
axs[2].plot(xf, 2.0/len(y_smooth) * np.abs(yf[0:len(y_smooth)//2]))
axs[2].set_title('Frequency Domain Signal after Smoothing')
axs[2].set_xlabel('Frequency [Hz]')
axs[2].set_ylabel('Amplitude')
axs[2].grid(linestyle = "--", color = "gray")

plt.tight_layout()
plt.show()
```
